```python
# ●	Write a Python program to show summary statistics of sum, mean, standard deviation, skewness, and kurtosis of any chosen variable.  
```

```python
# ●	Write a python program to find out top 15 jobs. Make a bar graph of salaries as well. 
```

```python
job_title=df["job_title"].unique();     #extract all unique value preswnt in job_title column
print("All unique job title.");
print(job_title);
print();
mean_salary_by_job = df.groupby("job_title")["salary"].mean();   #calculating average value of salary column according to categorized by  job_title
top_15_mean_salaries = mean_salary_by_job.nlargest(15);   #retrieve 15 high jobs salary value with index value
print();
print("top 15 job with highest average salary");
print(top_15_mean_salaries);
print();
print("top 15 job title");
top_15_job_title=top_15_mean_salaries.index;
print(top_15_job_title);
print();
print("top 15 job average salary");
top_15_job_salary=top_15_mean_salaries.values;
print(top_15_job_salary);
print();
#Bar graph of salary column by job_title
print("Bar graph of Maximum Salary by Job Title.");
plt.figure(figsize=(100, 100));                                                            #setting figure size
sb.barplot(x=df["job_title"],y=df["salary"],estimator='max',errorbar=('ci', 0));
plt.title("Maximum Salary by Job Title.");
plt.xlabel("Job Title");
plt.ylabel("Salary");
plt.xticks(rotation=45, ha='right');           #rotating label of x -axis by 45degree.
plt.subplots_adjust(left=0.2, right=0.8);  
st.use('ggplot');
plt.grid(color='y',linestyle='-',linewidth=1);         #setting style for grid
plt.show();
print();
top_salary_emp_row_details=df["salary"].nlargest(1);
print(top_salary_emp_row_details);
print();
top_salary_employee_details=df.loc[3669,:];
print("top_salary_employee_details");
print(top_salary_employee_details);
print("From above code we know Data Scientis have highest salary");
```

```python
# Write a python program to find out salaries based on experience level.Illustrate it through bar graph. 	
```

```python
print("types of experience level of employee");
experience_level=df["experience_level"].unique();
#display unique experience level of employee
print("display unique experience level of employee");
print(experience_level);
#calculating mean of salary group by experience level.
experience_salary=df.groupby(by="experience_level")["salary"].mean();
print();
print(experience_salary);
print();
#Experience level type
print("Experience level type");
experience_level_type=experience_salary.index;
print(experience_level_type);
print();
#Experience level type average  salary.
print("Experience level type average  salary.");
experience_level_type_avg_salary=experience_salary.values;
print(experience_level_type_avg_salary);
plt.figure(figsize=(16, 9));
sb.barplot(x=experience_level_type,y=experience_level_type_avg_salary,estimator='max',errorbar=('ci', 0));
plt.title("Average Salary according to experience level");
plt.xlabel("Experience level type");
plt.ylabel("Salary");
plt.xticks(rotation=45, ha='right');
st.use('ggplot');
plt.grid(color='y',linestyle='-',linewidth=1);
plt.show();
print("From above graph we know that Medium Level/Intermediate have high salary.");
```

```python
# ●	Write a Python program to show histogram and box plot of any chosen different variables. Use proper labels in the graph.   
```

```python
#selecting variables
print("Select variables");
variable_1="work_year";
variable_2="salary_in_usd";

#for variable 1
#histogram
width=0.2;
plt.figure(figsize=(16,9));
plt.subplot(2,2,1);
plt.hist(df[variable_1],bins=10,rwidth=width,histtype='bar',orientation='vertical',color='g',label='Work Year');
plt.title('Histogram of Work Year column');
plt.xlabel('Work Year');
plt.ylabel('No of occurence');
plt.legend(loc=2);
st.use('ggplot');
plt.grid(color='y',linestyle='--',linewidth=5);
plt.show();

#boxplot
plt.figure(figsize=(16,9));
plt.subplot(2,2,2);
sb.boxplot(x=df[variable_1],linewidth=2, color='lightgreen');
plt.title('Box Plot of Work Year column');
plt.xlabel('Work Year');
plt.show()

#for variable 2
#histogram
width=0.2;
plt.figure(figsize=(16,9));
plt.subplot(2,2,3);
plt.hist(df[variable_2],bins=10,rwidth=width,histtype='bar',orientation='vertical',color='g',label='Salary in USD');
plt.title('Histogram of Salary in USD column');
plt.xlabel('Salary in USD');
plt.ylabel('No of occurence');
plt.legend(loc=1);
st.use('ggplot');
plt.grid(color='y',linestyle='--',linewidth=5);
plt.show();

#boxplot
plt.figure(figsize=(16,9));
plt.subplot(2,2,4);
sb.boxplot(x=df[variable_2],linewidth=2, color='lightgreen');
plt.title('Box Plot of Salary in USD column');
plt.xlabel('Remote Ratio');
plt.show()
```
